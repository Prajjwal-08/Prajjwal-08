```python

import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
from scipy import stats
df = pd.read_csv (r'tips.csv')   #read the csv file (put 'r' before the path string to address any special characters in the path, such as '\'). Don't forget to put the file name at the end of the path + ".csv"
print (df)
```

#### Q.1 Find out Five-Number summary for "Total Bill" and "Tip"


```python
bill = df.total_bill
print("Five-Number summary for 'Total Bill'")
a=np.max(bill)
print("Maximum Bill = ",a)
b=np.min(bill)
print("Minimum Bill = ",b)
c=np.std(bill)
print("Standard Deviation = ",c)
d=np.median(bill)
print("Median =",d)
e=np.mean(bill)
print("Mean = ",e)
```

```python
Tipp = df.tip
print("Five-Number summary for 'Tip'")
print("Maximum Bill = ",np.max(Tipp))
print("Minimum Bill = ",np.min(Tipp))
print("Standard Deviation = ",np.std(Tipp))
print("Median =",np.median(Tipp))
print("Mean = ",np.mean(Tipp))
```

#### Q.4 Analyze the diagram to mark the skewness in the data


```python

plt.figure(figsize=[12,8])
plt.hist(x=bill, bins=20, color = 'g',edgecolor="g")
plt.xlabel("Total Bill",fontsize=19,color='red')
plt.ylabel("Distribution",fontsize=19,color='red')
plt.title('Normal Distribution Histogram For Total Bill - Bins 20',fontsize=19,color='red')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Q.8 Plot the cumulative Frequency Polygon for "Total Bill" and "Tip"

```python
values, base = np.histogram(bill, bins=20)
cumulative = np.cumsum(values)      
plt.plot(base[:-1], cumulative)

plt.title('Total_bills',fontsize=20,color='r')
plt.show()
```

```python

values, base = np.histogram(Tipp, bins=40)
cumulative = np.cumsum(values)
plt.plot(base[:-1], cumulative)
plt.title('Tip',fontsize=20,color='r')
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
plt.show()
```
